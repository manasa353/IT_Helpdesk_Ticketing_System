```python
# IT Helpdesk Ticketing System

# List to store tickets
tickets = []
ticket_id = 1

def add_ticket():
    global ticket_id
    title = input("Enter ticket title: ")
    description = input("Enter ticket description: ")
    ticket = {
        'id': ticket_id,
        'title': title,
        'description': description,
        'status': 'Open'
    }
    tickets.append(ticket)
    print("Ticket created successfully!")
    ticket_id += 1

def view_tickets():
    if not tickets:
        print("No tickets found.")
        return
    for ticket in tickets:
        print(f"ID: {ticket['id']}, Title: {ticket['title']}, Description: {ticket['description']}, Status: {ticket['status']}")

def update_ticket():
    try:
        tid = int(input("Enter ticket ID to update: "))
    except ValueError:
        print("Invalid input! Please enter a number.")
        return
    for ticket in tickets:
        if ticket['id'] == tid:
            new_status = input("Enter new status (Open/In Progress/Closed): ")
            if new_status in ['Open', 'In Progress', 'Closed']:
                ticket['status'] = new_status
                print(f"Ticket {tid} updated to status '{new_status}'")
            else:
                print("Invalid status!")
            return
    print("Ticket ID not found!")

def delete_ticket():
    try:
        tid = int(input("Enter ticket ID to delete: "))
    except ValueError:
        print("Invalid input! Please enter a number.")
        return
    for i, ticket in enumerate(tickets):
        if ticket['id'] == tid:
            tickets.pop(i)
            print(f"Ticket {tid} deleted.")
            return
    print("Ticket ID not found!")

def main():
    while True:
        print("\n--- IT Helpdesk Ticketing System ---")
        print("1. Add new ticket")
        print("2. View all tickets")
        print("3. Update ticket status")
        print("4. Delete a ticket")
        print("5. Exit")
        choice = input("Choose an option (1-5): ")
        
        if choice == '1':
            add_ticket()
        elif choice == '2':
            view_tickets()
        elif choice == '3':
            update_ticket()
        elif choice == '4':
            delete_ticket()
        elif choice == '5':
            print("Exiting program. Goodbye!")
            break
        else:
            print("Invalid choice. Please enter a number between 1 and 5.")

# Run the program
if __name__ == "__main__":
    main()
```
